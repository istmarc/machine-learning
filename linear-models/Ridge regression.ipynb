# Ridge regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("science")
```

## Dataset

```python
def f(x):
    return 1 + x + x**2
n = 20
x = np.arange(n) + np.random.randn()
y = f(x)
```

```python
fig = plt.subplots(figsize = (4, 4))
plt.scatter(x, y, color = "grey")
plt.title("Data")
plt.show()
```

## Ridge regression

regularized quadratic cost function

$$J(\Theta) = (y - X\Theta)^T (y - X\Theta) + \delta^2 \Theta^T \Theta$$

with

$$\Theta = [\theta_1, \ldots, \theta_d]$$

Ridge estimator

$$\hat{\Theta}_R = (X^TX + \delta^2 I_d)^{-1}X^Ty$$

```python
def ridge_estimator(X, y, delta, d):
    return np.linalg.inv(X.T @ X + delta**2 * np.identity(d)) @ X.T @ y
```

```python
d = 2
m = 5
deltas = np.linspace(1, 50, m)
deltas
```

## Simple model

```python
X = np.ones((n,2))
for i in range(n):
    X[i, 1] = x[i]
```

**OLS Estimator**

```python
est = ridge_estimator(X, y, 0., d)
est
```

```python
fig = plt.subplots(figsize = (4, 4))
plt.scatter(x, y, color = "grey")
model_ols = est[1] * range(n+1) + est[0]
plt.plot(model_ols)
plt.title("Simple model - OLS estimator")
plt.show()
```

**MSE of the OLS estimator**

```python
y_hat = X @ est
mse = 1/n * sum((y-y_hat)**2)
mse
```

**Ridge estimators**

```python
estimators = np.zeros((m, d))
for i in range(m):
    estimators[i] = ridge_estimator(X, y, deltas[i], d)
```

```python
estimators
```

```python
plt.plot(estimators[:, 0])
plt.plot(estimators[:, 1])
plt.title("Simple model - Regularization parths")
plt.show()
```

```python
fig = plt.subplots(figsize = (4, 4))
plt.scatter(x, y, color = "grey", label = "data")
for i in range(m):
    est = estimators[i]
    model = est[1] * range(n+1) + est[0]
    plt.plot(model, label = r"$\delta =$" + str(deltas[i]))
plt.legend()
plt.title("Simple model - Ridge estimators")
plt.show()
```

**Compute the MSE for each ridge estimator**

```python
mse = np.zeros(m)
for i in range(m):
    est = estimators[i]
    y_hat = X @ est
    mse[i] = 1/n * sum((y - y_hat)**2)
```

```python
mse
```

```python
plt.plot(mse)
plt.title("Mean Squared Error - Ridge estimators")
plt.show()
```

## Right model

```python
d = 3
X = np.ones((n,d))
for i in range(n):
    X[i, 1] = x[i]
    X[i, 2] = x[i]**2
```

```python
X
```

**OLS Estimator**

```python
est = ridge_estimator(X, y, 0., d)
est
```

```python
fig = plt.subplots(figsize = (4, 4))
plt.scatter(x, y, color = "grey", label = "data")
model_ols = np.zeros(n+1)
for i in range(n+1):
    model_ols[i] = est[2] * i**2 + est[1] * i + est[0]
plt.plot(model_ols, label = "model OLS")
plt.legend()
plt.title("Right model - OLS estimator")
plt.show()
```

**MSE of the OLS estimator**

```python
y_hat = X @ est
mse = 1/n * sum((y-y_hat)**2)
mse
```

**Ridge estimators**

```python
estimators = np.zeros((m, d))
for i in range(m):
    estimators[i] = ridge_estimator(X, y, deltas[i], d)
```

```python
estimators
```

```python
plt.plot(estimators[:, 0])
plt.plot(estimators[:, 1])
plt.plot(estimators[:, 2])
plt.title("Right model - Regularization paths")
plt.show()
```

```python
fig = plt.subplots(figsize = (4, 4))
plt.scatter(x, y, color = "grey", label = "data")
for i in range(m):
    est = estimators[i]
    model = np.zeros(n+1)
    for j in range(n+1):
        model[j] = est[2] * j**2 + est[1] * j + est[0]
    plt.plot(model, label = r"$\delta =$" + str(deltas[i]))
plt.legend()
plt.title("Right model - Ridge estimators")
plt.show()
```

**MSE for each ridge estimator**

```python
mse = np.zeros(m)
for i in range(m):
    est = estimators[i]
    y_hat = X @ est
    mse[i] = 1/n * sum((y - y_hat)**2)
```

```python
mse
```

```python
plt.plot(mse)
plt.title("Mean Squared Error - Ridge esstimators")
plt.show()
```

## Complex model

```python
d = 11
X = np.ones((n,d))
for i in range(n):
    for j in range(d):
        X[i, j] = x[i]**j
```

```python
X.shape
```

```python
X
```

**OLS estimator**

```python
est = ridge_estimator(X, y, 0., d)
est
```

```python
plt.plot(est)
plt.title("Complex model - OLS estimators")
plt.show()
```

```python
fig = plt.subplots(figsize = (4, 4))
plt.scatter(x, y, color = "grey", label = "data")
model_ols = np.zeros(n+1)
for i in range(n+1):
    for j in range(d):
        model_ols[i] += est[j] * i**j
plt.plot(model_ols, label = "model OLS")
plt.legend()
plt.title("Complex model - OLS estimator")
plt.show()
```

**MSE of the OLS estimator**

```python
y_hat = X @ est
mse = 1/n * sum((y-y_hat)**2)
mse
```

**Ridge estimators**

```python
estimators = np.zeros((m, d))
for i in range(m):
    estimators[i] = ridge_estimator(X, y, deltas[i], d)
```

```python
estimators
```

```python
for i in range(d):
    plt.plot(estimators[:, i])
plt.title("Complex model - Regularization paths")
plt.show()
```

```python
fig = plt.subplots(figsize = (4, 4))
plt.scatter(x, y, color = "grey", label = "data")
for i in range(m):
    est = estimators[i]
    model = np.zeros(n+1)
    for k in range(n+1):
        for j in range(d):
            model[j] += est[j] * k**j
    plt.plot(model, label = r"$\delta =$" + str(deltas[i]))
plt.legend()
plt.title("Complex model - Ridge estimators")
plt.show()
```

**MSE for each ridge estimator**

```python
mse = np.zeros(m)
for i in range(m):
    est = estimators[i]
    y_hat = X @ est
    mse[i] = 1/n * sum((y - y_hat)**2)
```

```python
mse
```

```python
plt.plot(mse)
plt.title("Mean Squared Error - Ridge estimators")
plt.show()
```
