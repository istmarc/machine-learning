# Energy demand prediction

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use("science")
```

## Dataset

```python
data = pd.DataFrame({"wind_speed": [100., 50., 45., 60.],
                     "people": [2, 42, 31, 35],
                     "energy": [5, 25, 22, 18]})
```

## Linear prediction

$$y = X\Theta$$

Let $y = [5., 25., 22., 18.]$, $X =\left[ \begin{array}{c c c}
                                     1 & 100 & 2\\
                                     1 & 50  & 42\\
                                     1 & 45  & 31\\
                                     1 & 60  & 35
                                     \end{array}
                                     \right]$
and $\Theta = [\theta_1, \theta_2, \theta_3]$

## Optimization approach

$$J(\Theta) = (y - X\Theta)^T(y - X\Theta)
            = \sum_{i = 1}^{n}(y_i - x_i^T\Theta)^2
            = \sum_{i=1}^{n}(y_i - \hat{y}_i)^2$$

```python
def objective_function(theta):
    return (y - X@theta).T @ (y - X@theta)
```

```python
y = data["energy"]
X = np.array([1., 100, 50., 1., 45, 60, 1., 2, 42, 1., 31, 35]).reshape(4, 3)
```

```python
theta_2 = np.linspace(-2, 2)
theta_3 = np.linspace(-2, 2)
```

**Contour plot**

```python
obj = np.zeros((theta_2.size, theta_3.size))
for i in range(theta_2.size):
    for j in range(theta_3.size):
        obj[i,j] = objective_function(np.array([1., theta_2[i], theta_3[j]]))
```

```python
fig = plt.subplots(figsize=(4, 4))
plt.contour(obj)
plt.title("Contour plot")
plt.show()
```

**Surface plot**

```python
t2_mesh, t3_mesh = np.meshgrid(theta_2, theta_3)
```

```python
fig, ax = plt.subplots(subplot_kw = {"projection": "3d"}, figsize = (5, 5))
ax.plot_surface(t2_mesh, t3_mesh, obj, cmap = plt.cm.coolwarm)
ax.contour(t2_mesh, t3_mesh, obj, extend3d=False, zdir="z", offset=0.)
plt.title("Surface plot")
plt.show()
```

## Optimization

$$J(\Theta) = (y - X\Theta)^T (y - X\Theta)$$

$$\frac{dJ(\theta)}{d\Theta} = \frac{d}{d\Theta} [y^Ty -y^TX\Theta - \Theta^T X^Ty + \Theta^TX^TX\Theta]$$

$$\frac{dJ(\Theta)}{d\Theta} = -2X^Ty + 2 XX^T\Theta$$

$dJ(\Theta)/d\Theta = 0$ gives the normal equation $X^TX\Theta = X^Ty$

The least square estimate of $\Theta$ is 
$$\hat{\Theta} = (X^TX)^{-1}X^Ty$$

```python
theta_hat = np.linalg.inv(X.T @ X) @ X.T @ y
```

```python
theta_hat
```

## Predictions on the training set

```python
fig = plt.subplots(figsize=(5, 4))
plt.plot(y, label = "data y")
y_hat = X@theta_hat
plt.plot(y_hat, label = "prediction $\hat{y}$")
plt.scatter(range(y.size), y)
plt.scatter(range(y_hat.size), y_hat)
plt.legend()
plt.show()
```
